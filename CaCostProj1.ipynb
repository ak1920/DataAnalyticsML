```python
df['Year'] = df['Year'].astype('float64')
```

```python
X2 = DataFrame(df, columns=['Year'])
y2 = DataFrame(df, columns=['Total Costs'])
```

```python
regression = LinearRegression()
regression.fit(X2,y2)
```

```python
regression.coef_
```

```python
regression.intercept_
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X2,y2, alpha=0.3)
plt.plot(X2, regression.predict(X2), color='red')

plt.title('Total Costs of Cancer per Year')
plt.xlabel('Year')
plt.ylabel('Total Costs')
plt.xlim(2009,2021)
plt.ylim(0,300000)
plt.show()
```
